This problem asks us to:

Write code that does the following: 

opens the number_list.txt file that was created by the code you wrote in question 3, reads all of the numbers from the file and displays them, then closes the file.

To complete this problem I am going to have to use the open function again to open the file, but this time I am going to have to use the read function to read its contents. I am also going to have to use a loop again to display the text within the file line by line. 

To start I am going to once again establish a main function and open the file a in read mode. 

```python
def main():
    contents = open('number_list.txt','r')
    
    contents.close()
if __name__ == '__main__':
    main()   
```

It did not throw an error so I think it opened and closed the file without issue. Now to add in the part that displays the contents within the file. 

```python
def main():
    
    contents = open('number_list.txt','r')
    
    for line in contents:
        print(line)
    
    contents.close()
if __name__ == '__main__':
    main()   
```

That has worked, however there is a space in between each number so I am going to add a 'strip' function to remove the '\n' from each of the numbers. 

```python
#Define the main function
def main():
    
    #Open the "number_list.txt" file in read mode and assign it to the contents variable. 
    contents = open('number_list.txt','r')
    
    #for every line in the 'contents' file repeat the code below
    for line in contents:
        #This displays the current line in the file with the '\n' removed from the sting so that it does not add an extra line
        print(line.rstrip('\n'))
    
    #Close the 'contents' file
    contents.close()
    
#if the file name contains '__main__' run the 'main' function
if __name__ == '__main__':
    #Run the main function
    main()    
```

There we go! That look good I think this problem is done!

```python
def main():
    print("Enter a file name in the box below and its contents will be displayed with numbered lines.")
    user_file = input('File name: ')
    
    try:
        with open(user_file,'r') as file:
            count = 1
        
            for line in file:
                print(f'{count}: {line}')
                count += 1
                
    except FileNotFoundError:
        print(f'The file {user_file} could not be found.')
    except:
        print("An error has occured!")
    
if __name__ == '__main__':
    main()  
```
