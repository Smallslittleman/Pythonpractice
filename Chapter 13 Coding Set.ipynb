```python
def main():
    class main_gui:
        def __init__(self):
            self.main_window = tkinter.Tk()
            
            self.main_window.title("Name and Address Program")
            
            self.top_frame = tkinter.Frame()
            self.middle_frame = tkinter.Frame()
            self.bottom_frame = tkinter.Frame()
            
            self.directions = tkinter.Label(self.top_frame, 
                                            text='Press the info button for info, and the close button to close the program.')
            
            self.directions.pack()
            
            self.top_frame.pack()
            
            
            
            
            
            self.info_button = tkinter.Button(self.bottom_frame, text = "Press for info")
            self.close_button = tkinter.Button(self.bottom_frame, text = "Close", command = self.main_window.destroy) 
            
            self.info_button.pack(side = 'left')
            self.close_button.pack(side ='right')
            
            self.bottom_frame.pack()
            
            
            tkinter.mainloop()
    
    main_gui()
        
if __name__ == '__main__':
    main()
```

The close button works, now to add the labels which give the info and get the info button working!

```python
#import the tkinter module
import tkinter

#define a main function
def main():
    #define the main_gui class
    class main_gui:
        #define the __init__ method within the main_gui class
        def __init__(self):
            #make a main window
            self.main_window = tkinter.Tk()
            #title the main window to the text below
            self.main_window.title("Name and Address Program")
            
            #create a top frame
            self.top_frame = tkinter.Frame()
            #create a middle frame
            self.middle_frame = tkinter.Frame()
            #create a bottom frame
            self.bottom_frame = tkinter.Frame()
            
            #TOP FRAME
            # ----------------------------------------------------------------------------------------------------------------
            
            #create a label which displays some directions for the user
            self.directions = tkinter.Label(self.top_frame, 
                                            text='Press the info button for info, and the close button to close the program.')
            
            #pack the directions label into the top frame
            self.directions.pack()
            
            #pack the top frame into the main window
            self.top_frame.pack()
            
            # ----------------------------------------------------------------------------------------------------------------
            
            #MIDDLE FRAME
            
            # ----------------------------------------------------------------------------------------------------------------
            
            #create a variable called name which is a string variable
            self.name = tkinter.StringVar()
            #create a variable called address which is a string variable
            self.address = tkinter.StringVar()
            #create a variable called town which is a string variable 
            self.town = tkinter.StringVar()
            
            #create a name label which is located in the middle frame and its text is set to the name variable
            self.name_label = tkinter.Label(self.middle_frame, textvariable = self.name)
            #create a address label which is located in the middle frame and its text is set to the address variable 
            self.address_label = tkinter.Label(self.middle_frame, textvariable = self.address)
            #create a town label which is located in the middle frame and its text is set to the address variable
            self.town_label = tkinter.Label(self.middle_frame, textvariable = self.town)
            
            #pack the name label into the middle frame
            self.name_label.pack()
            #pack the address label into the middle frame
            self.address_label.pack()
            #pack the town label into the middle frame
            self.town_label.pack() 
            
            #pack the middle frame into the main window
            self.middle_frame.pack()
            
            # ----------------------------------------------------------------------------------------------------------------
            
            #BOTTOM FRAME
            
            # ----------------------------------------------------------------------------------------------------------------
            
            #create a info button which is in the bottom frame and its command changes the info labels text
            self.info_button = tkinter.Button(self.bottom_frame, text = "Press for info", command = self.set_info)
            #create a close button which is in the bottom frame and its command destroys the main window
            self.close_button = tkinter.Button(self.bottom_frame, text = "Close", command = self.main_window.destroy) 
            
            #pack the info button into the bottom frame on the left side
            self.info_button.pack(side = 'left')
            #pack the close button into the bottom frame on the right side
            self.close_button.pack(side ='right')
            
            #pack the bottom frame into the main window
            self.bottom_frame.pack()
            
            # ----------------------------------------------------------------------------------------------------------------
            
            #run the main tkinter loop
            tkinter.mainloop()
        
        #define the set_info method
        def set_info(self):
            #set the name variable to the name
            self.name.set("David Turner")
            #set the address variable to the address
            self.address.set("432 Jamesmark Rd")
            #set the town variable to the town
            self.town.set("Ivantown, MI 76609")
            
    #run the main_gui class
    main_gui()
#if the name equals __main__ then run the main function        
if __name__ == '__main__':
    #run the main function
    main()
```
