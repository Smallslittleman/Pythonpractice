```python
#Define the tuple with the course numbers
course_numbers = ('CS101','CS102','CS103','NT110','CM241')
#Define the tuple with the room numbers
room_numbers = ('3004','4501','6755','1244','1411')
#define the tuple with the instructors names
instructors = ('Haynes','Alvarado','Rich','Burke','Lee')
#define the tuple with the meeting times for each class 
meeting_times = ('8:00a.m.','9:00a.m.','10:00a.m.','11:00a.m.','1:00p.m.')

#define the main function 
def main():
    #call the coursesmod module and the directions function within it.
    coursesmod.directions()
    #make a blank dictionary called 'rm_dict'
    rm_dict = {}
    #make a blank dictionary called 'instruct_dict'
    instruct_dict = {}
    #make a blank dictionary called 'mt_dict'
    mt_dict = {}
    #reset the count variable to 0
    count = 0 
    
    #repeat for each item in the course_numbers tuple
    for item in course_numbers:
        #add an item to the rm_dict with the current item in the course numbers list being set to the key and the current room
        #number being set to the value
        rm_dict[item] = room_numbers[count]
        #add an item to the instruct_dict with the current item in the course numbers list being set to the key and the current
        #instructors name being set to the value
        instruct_dict[item] = instructors[count]
        #add an item to the mt_dict with the current item in the course numbers list being set to the key and the current
        #meeting time being set to the value
        mt_dict[item] = meeting_times[count]
        
        #add 1 to the count variable each time the loop repeats 
        count += 1
    
    #assign the variable 'user' to the users entery 
    user = input('Type your course number here: ')
    
    #print the room number that corresponds to the course number entered
    print(f'The room number is: {rm_dict.get(user,"INVALID ROOM NUMBER")}')
    #print the instructor name that corresponds to the course number entered
    print(f'The instructor is: {instruct_dict.get(user,"INVALID ROOM NUMBER")}')
    #print the meeting time that corresponds to the course number entered
    print(f'The meeting time is: {mt_dict.get(user,"INVALID ROOM NUMBER")}')
       
#if the file name contains '__main__' run the 'main' function    
if __name__ == '__main__':
    #Call the main function
    main()  
```
